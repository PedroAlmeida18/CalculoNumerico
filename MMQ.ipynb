# MMQ

```python
import matplotlib.pyplot as plt
import numpy as np

def mmq(g1, g2, x, y):

  a_1 = 0
  a_2 = 0

  G_11 = np.dot(g_1(x), g_1(x))
  G_12 = np.dot(g_1(x), g_2(x))
  G_21 = G_12
  G_22 = np.dot(g_2(x), g_2(x))
  G = [[G_11, G_12],
      [G_21, G_22]]

  b_1 = np.dot(y,g_1(x)) # Faz o produto interno
  b_2 = np.dot(y,g_2(x))


  b = [b_1,b_2]

  a = np.linalg.solve(G, b)

  a_1 = a[0]
  a_2 = a[1]

  return a_1, a_2

def g_1(r):
    return np.ones(len(r))

def g_2(r):
    return 1/(r**2)


x = np.array([1, 1.5, 2, 2.5, 3])
y = np.array([1.6, 5.6, 6, 7.1, 7])

a_1, a_2 = mmq(g_1,g_2,x,y)
# Defina a função de ajuste:
def phi(z):
    return a_1 + a_2*g_2(z)



print("ALUNO : PEDRO HENRIQUE DE ALMEIDA SANTOS")
x_r = np.linspace(min(x), max(x), 100)
y_r = phi(x_r)
plt.plot(x, y, 'o')
plt.plot(x_r, y_r)
plt.show()

erro = 0
n = len(y)
for i in range (0,n):
  erro+= (y[i] - phi(x[i]))**2
print(f"Primeira possui um erro de aproxiamdamente {erro} \n")


def g_1(r):
    return np.ones(len(r))

def g_2(r):
    return np.log(r)


x = np.array([1, 1.5, 2, 2.5, 3])
y = np.array([1.6, 5.6, 6, 7.1, 7])


a_1, a_2 = mmq(g_1,g_2,x,y)

def phi(z):
    return a_1 + a_2*g_2(z)



x_r = np.linspace(min(x), max(x), 100)
y_r = phi(x_r)
plt.plot(x, y, 'o')
plt.plot(x_r, y_r)
plt.show()

erro = 0
n = len(y)
for i in range (0,n):
  erro+= (y[i] - phi(x[i]))**2
print(f"Segunda possui um erro de aproxiamdamente {erro} \n")

```
