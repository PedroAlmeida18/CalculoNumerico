# MMQ

```python
import numpy as np
import matplotlib.pyplot as plt

def mmq(g1, g2, x, y):

  g_1 = []
  g_2 = []

  for i in range(len(x)):
    g_1.append(g1(x[i]))
    g_2.append(g2(x[i]))


  G_11 = np.dot(g_1, g_1)
  G_12 = np.dot(g_1, g_2)
  G_22 = np.dot(g_2, g_2)

  A = [[G_11, G_12],[G_12, G_22]]

  vetor = [np.dot(y,g_1), np.dot(y,g_2)]

  a = np.linalg.solve(A, vetor)
  return a[0], a[1]


def g1(r):
  return 1

def g2(r):
  return (1/(r**2))

def g3(x):
  return np.log(x)

# condições iniciais

a1 = 0
a2 = 0

x = np.array([1, 1.5, 2, 2.5, 3])
y = np.array([1.6, 5.6, 6, 7.1, 7])

a1,a2 = mmq(g1, g2, x, y)

def phi(x):
  return a1*g1(x) + a2 * g2(x)


# Cálculo do espaço linear
x_r = np.linspace(min(x), max(x), 100)
y_r = phi(x_r)

plt.plot(x , y ,"o")
plt.plot(x_r , y_r, label= "")

# Calculando o erro
print("Aluno : PEDRO HENRIQUE DE ALMEIDA SANTOS ")
Erro = 0

for i in range(len(x)):
  Erro += (y[i] - phi(x[i]))**2
print(f"A primeira função gera um gráfico com um erro de aproxiamdamente {Erro} \n" )


def phi2(x):       # função do erro para a segunda equação dada.
  return a1*g1(x) + a2 * g3(x)

a1,a2 = mmq(g1, g3, x, y)
y_r = phi2(x_r)

plt.plot(x_r , y_r)
plt.legend(["pontos", "1° Função", "2° Função"])
plt.show

Erro = 0

for i in range(len(x)):
  Erro += (y[i] - phi2(x[i]))**2

print(f"já a segunda possui um erro de aproxiamdamente {Erro} \n" )

```
