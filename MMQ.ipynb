# MMQ

```python
import matplotlib.pyplot as plt
import numpy as np

def mmq(g1, g2, x, y):

  a_1 = 0
  a_2 = 0

  G_11 = np.dot(g_1(x), g_1(x))
  G_12 = np.dot(g_1(x), g_2(x))
  G_21 = G_12
  G_22 = np.dot(g_2(x), g_2(x))
  G = [[G_11, G_12],
      [G_21, G_22]]

  b_1 = np.dot(y,g_1(x))
  b_2 = np.dot(y,g_2(x))


  b = [b_1,b_2]

  a = np.linalg.solve(G, b)

  a_1 = a[0]
  a_2 = a[1]

  return a_1, a_2

def g_1(r):
    return np.ones(len(r))

def g_2(r):
    return 1/(r**2)


x = np.array([1, 1.5, 2, 2.5, 3])
y = np.array([1.6, 5.6, 6, 7.1, 7])

a_1, a_2 = mmq(g_1,g_2,x,y)

def phi(z):
    return a_1 + a_2*g_2(z)




x_r = np.linspace(min(x), max(x), 100)
y_r = phi(x_r)
plt.title("Gráficos da Primeira  Função - Pedro Henrique de Almeida Santos ")
plt.plot(x, y, 'o')
plt.plot(x_r, y_r)
plt.show()

erro = 0
n = len(y)
for i in range (0,n):
  erro+= (y[i] - phi(x[i]))**2
print(f"Primeira possui um erro de aproxiamdamente {erro} \n")


def g_1(r):
    return np.ones(len(r))

def g_2(r):
    return np.log(r)


x = np.array([1, 1.5, 2, 2.5, 3])
y = np.array([1.6, 5.6, 6, 7.1, 7])


a_1, a_2 = mmq(g_1,g_2,x,y)

def phi2(z):
    return a_1 + a_2*g_2(z)



x_r = np.linspace(min(x), max(x), 100)
y_r = phi(x_r)
plt.title("Gráficos da Segunda Função - Pedro Henrique de Almeida Santos ")
plt.plot(x, y, 'o')
plt.plot(x_r, y_r)
plt.show


erro = 0
n = len(y)
for i in range (0,n):
  erro+= (y[i] - phi(x[i]))**2

print(f"Segunda possui um erro de aproxiamdamente {erro} \n")
```

```python

import matplotlib.pyplot as plt
import numpy as np
print("GRÁFICO COM AS DUAS FUNÇÕES :")

def mmq(g1, g2, x, y):
    g_1 = []
    g_2 = []

    for i in range(len(x)):
        g_1.append(g1(x[i]))
        g_2.append(g2(x[i]))

    G_11 = np.dot(g_1, g_1)
    G_12 = np.dot(g_1, g_2)
    G_22 = np.dot(g_2, g_2)

    A = [[G_11, G_12], [G_12, G_22]]

    vetor = [np.dot(y, g_1), np.dot(y, g_2)]

    a = np.linalg.solve(A, vetor)
    return a[0], a[1]

def g1(r):
    return 1

def g2(r):
    return (1 / (r ** 2))

def g3(x):
    return np.log(x)

x = np.array([1, 1.5, 2, 2.5, 3])
y = np.array([1.6, 5.6, 6, 7.1, 7])

a1, a2 = mmq(g1, g2, x, y)

def phi(x):
    return a1 * g1(x) + a2 * g2(x)

x_r = np.linspace(min(x), max(x), 100)
y_r = phi(x_r)

plt.plot(x, y, "o")
plt.plot(x_r, y_r, label="1ª Função")

erro1 = 0
for i in range(len(x)):
    erro1 += (y[i] - phi(x[i])) ** 2

print(f"A primeira função gera um gráfico com um erro de aproximadamente {erro1} \n")


def phi2(x):
    return a1 * g1(x) + a2 * g3(x)


a1,a2 = mmq(g1, g3, x, y)
y_r = phi2(x_r)


plt.plot(x_r, y_r, label="2ª Função")

erro2 = 0
for i in range(len(x)):
    erro2 += (y[i] - phi2(x[i])) ** 2

print(f"Já a segunda possui um erro de aproximadamente {erro2} \n")

plt.legend(["Pontos", "1ª Função", "2ª Função"])
plt.xlabel("x")
plt.ylabel("y")
plt.title("Gráficos das Funções")
plt.show()
```
