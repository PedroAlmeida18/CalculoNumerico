```python
import pandas  as pd
import numpy as np
import matplotlib.pyplot as plt
def interpNewton(x,y,xi):
  n = len(x)
  fdd = [[None for x in range(n)] for x in range(n)]
  yint = 0
  for i in range(n):
    fdd[i][0]= y[i]

  for j in range (1,n):
    for i in range(n-j):
      fdd[i][j]= (fdd[i+1][j-1] - fdd[i][j-1])/(x[i+j]-x[i])

  #fdd_table =pd.DataFrame(fdd)
  #print(fdd_table)
  xterm = 1
  yint= fdd[0][0]
  for order in range(1,n):
    xterm = xterm*(xi-x[order-1])
    yint= yint + fdd[0][order]*xterm


  return  yint

x = [30,40, 45]
y = [0.99818,0.99828, 0.99849]
xp = float(input("Digite o valor de uma temperatura em Graus Celsius entre 30 e 40 para descobrir seu calor específico:"))
yp = interpNewton(x,y,xp)

xp_aux = np.linspace(30,45,100)

yp_aux = []

for j in range(0,len(xp_aux)):

  yp_aux.append(interpNewton(x,y,xp_aux[j]))
print("Atividade 04 - Interpolção polinomial, pelo método de Newton ")
print("Aluno : Pedro Henrique de Almeida Santos")
print ( "O CALOR ESPECIFICO DE", xp," GRAUS CELSIUS É :" ,yp)
print("Gráfico de calor específico pela temperatura em Celsius")
t = np.arange(-1,-1,-1)
yt=[]
for i in t :
  yt.append(interpNewton(x,y,i))
plt.plot(t,yt,'b-')
plt.plot(x,y,"ro")
plt.plot(xp,yp,'g*')
plt.plot(xp_aux, yp_aux)
plt.grid()
plt.show()
```
